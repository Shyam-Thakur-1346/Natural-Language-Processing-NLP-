```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import string
```

```python
df=pd.read_csv("train.txt",sep=";",header=None,names=['text','emotions'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
unique_emotions = df['emotions'].unique()
```

```python
unique_emotions
```

```python
emotion_numbers={}
i=0
for emo in unique_emotions:
    emotion_numbers[emo]=i
    i+=1
df["emotions"]=df["emotions"].map(emotion_numbers)
```

```python
df.head()
```

doing all text in lowercase

```python
df["text"]=df["text"].apply(lambda x: x.lower())
```

removing punctuations (special character)

```python
import string
def remove_punc(txt):
    return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df['text'] = df['text'].apply(remove_punc)
```

```python
def remove_numbers(txt):
    new = ""
    for i in txt:
        if not i.isdigit():
            new = new + i
    return new

df['text'] = df['text'].apply(remove_numbers)
```

```python
def remove_emojis(txt):
    new = ""
    for i in txt:
        if i.isascii():
            new += i
    return new

df['text'] = df['text'].apply(remove_emojis)
```
