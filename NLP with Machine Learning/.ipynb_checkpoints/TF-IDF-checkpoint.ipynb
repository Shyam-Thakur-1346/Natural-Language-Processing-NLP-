```python
# TF-IDF - (Term Frequency) ( Inverse Document Frequency)
```

```python
# formula TF - (no of occurence of term in document)/(total no of terms in documents)
```

```python
# formula IDF - log e(Total no of doc in corpus)/(no of doc with terms in them)
```

```python
# final = TF*IDF
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents =[
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]
```

```python
vectorizer=TfidfVectorizer()
```

```python
X=vectorizer.fit_transform(documents)
```

```python
print("Vocabulary:",vectorizer.get_feature_names_out())
print("\nTdidf Matrix:\n",X.toarray())
```

```python
# pros - information retrival (helps in google search)
# cons - sparsity , out of vacabulary , dimensions
```
