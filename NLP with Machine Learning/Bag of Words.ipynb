```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents =[
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]
```

```python
vectorizer=CountVectorizer(ngram_range=(3,3)) # ngrams, bigrams, trigrams ,ngrams
```

```python
X=vectorizer.fit_transform(documents)
```

```python
print("Vocabulary:",vectorizer.get_feature_names_out())
print("\nBow Matrix:\n",X.toarray())
```

```python
# ngram can be callsed as bag of ngrams
# pros - semantic meaning captured little bit
# cons- dimensionality increment , out of vocabulary
```
